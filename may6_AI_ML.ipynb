```python
import numpy as np
import pandas as pd
data=pd.read_csv("/content/student_placement_dataset (3).csv")
data.head(6)
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
print(data.columns)
```

```python
data.dropna(subset=['Placed'])
```

```python
data['Certifications']=data['Certifications'].map({'No':0,'Yes':1})
```

```python
data['Placed']=data['Placed'].map({'No':0,'Yes':1})
```

```python
data.head()
```

```python
data.loc[6]
```

```python
data.isna().sum()
```

```python
data=pd.get_dummies(data,columns=['Communication_Skills','Coding_Skills'],drop_first=True,dtype=int)
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
y=data['Placed']
x=data.drop(['Placed'],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape,x_test.shape,x_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score




```

```python
scaler=StandardScaler()
x_train_scled=scaler.fit_transform(x_train)
x_test_scled=scaler.transform(x_test)
```

```python

logistic_model=LogisticRegression()
logistic_model.fit(x_train_scled,y_train)
y_pred=logistic_model.predict(x_test_scled)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```

```python
new_data = pd.DataFrame({
    'CGPA': [4.5],
    'Internships': [0],
    'Aptitude_Score': [85],
    'Projects_Completed': [0],
    'Attendance_Percentage': [90],
    'Certifications': [0],
    'Communication_Skills_Average': [0],
    'Communication_Skills_Excellent': [0],
    'Communication_Skills_Good': [0],
    'Communication_Skills_Poor': [1],
    'Coding_Skills_Beginner': [1],
    'Coding_Skills_Intermediate': [0]
})
```

```python
new_data=new_data[x.columns]
```

```python
new_data_scaled=scaler.transform(new_data)
```

```python
prediction=logistic_model.predict(new_data_scaled)
probability=logistic_model.predict_proba(new_data_scaled)

print(f"prediction for new data:{"placed" if prediction[0]==1 else 'not placed'}")
print(f"probability of placement:{probability[0][1]:.2f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
metric_k=[]
neighbors=np.arange(3,15)
for k in neighbors:
  knn_classifier=KNeighborsClassifier(n_neighbors=k,metric='minkowski')
  knn_classifier.fit(x_train,y_train)
  y_pred=knn_classifier.predict(x_test)
  accuracy=accuracy_score(y_test,y_pred)
  metric_k.append(accuracy)
metric_k
```

```python
confusion_matrix(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k value')
plt.ylabel('Accuracy')
```

```python
knn_classifier=KNeighborsClassifier(n_neighbors=9,metric='minkowski')
knn_classifier.fit(x_train,y_train)
y_pred=knn_classifier.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
prediction=knn_classifier.predict(new_data)
print(prediction)
print(f"prediction for new data:{'placed' if prediction[0]==1 else 'not placed'}")
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier(max_depth=5,random_state=42)
dt_model.fit(x_train,y_train)
y_pred=dt_model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
prediction=dt_model.predict(new_data)
print(prediction)
print(f"prediction for new data:{'placed' if prediction[0]==1 else 'Not placed'}")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=150,max_depth=10,min_samples_split=20,random_state=42)
rf_model.fit(x_train,y_train)
y_pred = rf_model.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
prediction=rf_model.predict(new_data)
print(prediction)
print(f"prediction for new data:{'placed' if prediction[0]==1 else 'Not placed'}")
```
